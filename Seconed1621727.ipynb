```python
# libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
from IPython.display import Image
%matplotlib inline
```

```python
#Convert image to BGR in OpenCV
syrianCoin = cv2.imread(path+'SyrianCoin.jpg',cv2.IMREAD_COLOR)
coinRGB = cv2.cvtColor(syrianCoin, cv2.COLOR_BGR2RGB)

#Convert to  Gray
coinGray = cv2.cvtColor(syrianCoin, cv2.COLOR_BGR2GRAY)

plt.imshow(syrianCoin)
plt.imshow(coinGray, cmap='gray')
```

```python
#Convert image to Grayscale 
syrianCoin = cv2.imread(path+"SyrianCoin.jpg")
gray = cv2.cvtColor(syrianCoin, cv2.COLOR_BGR2GRAY)
plt.imshow(coinGray, cmap='gray')
```

```python
# GaussianBlur Effect 
 
blurCoin = cv2.GaussianBlur(coinGray, (11,11),0) 
plt.imshow(blurCoin, cmap="gray")
```

```python
#Aplly Canny filter

cannyCoin = cv2.Canny(blurCoin, 30, 300, 3)
plt.imshow(cannyCoin, cmap='gray')
```

```python
#Increase Edges Thickness

dilatedCoin = cv2.dilate(cannyCoin, (1,1), iterations=2)
plt.imshow(dilatedCoin, cmap='gray')
```

```python
#Contours 
(cnt, heirarchy) = cv2.findContours(dilatedCoin.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# the thickness which is the last parameter
rgbCoin = cv2.cvtColor(syrianCoin, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgbCoin, cnt, -1, (0,255,0), 2) 

plt.imshow(rgbCoin)
```

```python
#Print the counter value
print("The number of coins in the image is : ", len(cnt))
```
